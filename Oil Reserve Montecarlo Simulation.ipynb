```python
import numpy as np
import matplotlib.pyplot as plt

def ooip():
    a = 26880 # a is in acres
    h = np.random.uniform(750, 900) # h is in ft
    phi = np.random.uniform(0.08, 0.11) # this is a percentage
    sw = np.random.uniform(0.51, 0.67) # this is a percentage
    Bo = np.random.uniform(1.55, 1.58) # Bo is bbl/STB

    ooip = (7758* a* h* phi* sw)/ Bo
    return ooip

hist = []
it = 0
while it < 100000:
    hist.append(ooip())
    it += 1
print('The P90 value of the reserves is:', np.percentile(hist, 10))
fig = plt.figure(figsize = (10,6))
ax = fig.subplots()

ax.hist(hist, bins = 1000, density = True, cumulative = -1, histtype = 'step')
ax.axhline(0.9, color = 'r')
ax.axhline(0.5, color = 'r')
ax.axhline(0.2, color = 'r')

ax.grid(True)
ax.set_title('Cumulative step histograms')
ax.set_xlabel('OOIP STB')
ax.set_ylabel('Likelihood of occurrence')
plt.show()
```
